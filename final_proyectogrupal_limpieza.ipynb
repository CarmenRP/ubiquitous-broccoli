```python
def categorize_variable(col):
    if col.dtype == 'O':
        if col.nunique() == 2:
            return 'Binaria'
        elif col.name.lower().startswith('id') or col.name.lower().endswith('id'):
            return 'Índice de Nombres'
        else:
            return 'Categorica Nominal'
    elif pd.api.types.is_integer_dtype(col) or pd.api.types.is_float_dtype(col):
        if col.nunique() == len(col):
            if col.name.lower().startswith('id') or col.name.lower().endswith('id'):
                return 'Índice Numérico'
            else:
                return 'Categorica Nominal'
        elif col.nunique() < 20:
            return 'Numerica Discreta'
        else:
            return 'Numerica Continua'
    elif pd.api.types.is_categorical_dtype(col):
        if col.ordered:
            return 'Categorica Ordinal'
        else:
            return 'Categorica Nominal'
    else:
        return 'Desconocido'

def create_summary_table(df):
    summary = pd.DataFrame({
        'Tipo de Dato': df.dtypes,
        'Cardinalidad': df.nunique(),
        '% Cardinalidad': (df.nunique()/len(df))*100,
        'Valores Faltantes': df.isnull().sum(),
        '% Valores Faltantes': (df.isnull().sum()/len(df))*100,
    })

    summary['Categoría'] = df.apply(categorize_variable)
    
    return summary
```

```python
create_summary_table(df)
```

```python
def descriptive_statistics(df):
    num_stats = df.select_dtypes(include=[np.number]).describe().T
    cat_stats = df.select_dtypes(exclude=[np.number]).describe(include='all').T
    
    # Calcular estadísticas adicionales para variables numéricas
    num_stats['skewness'] = df.select_dtypes(include=[np.number]).skew()
    num_stats['kurtosis'] = df.select_dtypes(include=[np.number]).kurtosis()
    num_stats['median'] = df.select_dtypes(include=[np.number]).median()
    num_stats['mode'] = df.select_dtypes(include=[np.number]).mode().iloc[0]
    num_stats['iqr'] = num_stats['75%'] - num_stats['25%']
    num_stats['range'] = num_stats['max'] - num_stats['min']
    num_stats['variance'] = df.select_dtypes(include=[np.number]).var()
    num_stats['std_dev'] = df.select_dtypes(include=[np.number]).std()
    num_stats['missing'] = df.select_dtypes(include=[np.number]).isnull().sum()
    num_stats['missing_percent'] = df.select_dtypes(include=[np.number]).isnull().mean() * 100
    
    # Calcular estadísticas adicionales para variables categóricas
    cat_stats['missing'] = df.select_dtypes(exclude=[np.number]).isnull().sum()
    cat_stats['missing_percent'] = df.select_dtypes(exclude=[np.number]).isnull().mean() * 100
    
    # Ordenar columnas
    num_columns_order = ['count', 'mean', 'std', 'min', '25%', '50%', '75%', 'max', 
                         'median', 'mode', 'iqr', 'range', 'variance', 'std_dev', 
                         'skewness', 'kurtosis', 'missing', 'missing_percent']
    cat_columns_order = ['count', 'unique', 'top', 'freq', 'missing', 'missing_percent']
    
    num_stats = num_stats[num_columns_order]
    cat_stats = cat_stats[cat_columns_order]
    
    return num_stats, cat_stats
```

```python
num, cat = descriptive_statistics(df)
```
